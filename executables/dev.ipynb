```python
import pandas as pd
import plotly.express as px
from pathlib import Path
```

```python
data_path = Path('').resolve().parent / 'data' / 'panel_input_output.csv'
```

```python
df = pd.read_csv(data_path, parse_dates=['dt'])
```

```python
px.line(
    df,
    x='dt',
    y=['target', 'pred_init', 'pred'],
    animation_frame='obj_id'
)
```

```python
px.line(
    df.groupby(['segment', 'dt'], as_index=False)[['target', 'pred_init', 'pred']].sum(),
    x='dt',
    y=['target', 'pred_init', 'pred'],
    animation_frame='segment'
)
```
